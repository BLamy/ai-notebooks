# Set Open AI API key from ENV variables

```python
import os
import openai
openai.api_key = os.environ["OPENAI_API_KEY"]
```

# Creates a Prompt using Lang chain

```python
from langchain import PromptTemplate
template = """I want you to act as a joke generator
Make sure the jokes are funny and have some kind of pun.
Can you tell me a {jokeType} Joke?
"""

prompt = PromptTemplate(
    input_variables=["jokeType"],
    template=template,
)
```

# Creates a Web UI with gradio 

```python
import gradio

def generateJoke(jokeType):
    res = openai.Completion.create(
      model="text-davinci-003",
      prompt=prompt.format(jokeType=jokeType),
      temperature=0.7,
      max_tokens=256,
      top_p=1,
      frequency_penalty=0,
      presence_penalty=0
    )
    return res.choices[0].text

gradio.Interface(generateJoke, gradio.Radio(["funny", "dumb", "dad", "anti"]), "text").launch(share=True)
```
